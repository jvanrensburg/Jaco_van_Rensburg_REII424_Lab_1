```python
import csv
import pandas as pd
import numpy as np
import numpy.random
import datetime
filename = "appletaste.csv"
f=open(filename,"r")
df = pd.read_csv(f)
from matplotlib import pyplot as plt
#pie chart
applelist = df.groupby('panelist').sum()
var = 'aftertaste'
applelist[var].plot(kind='pie', fontsize=20)
plt.ylabel(var, horizontalalignment='left')
plt.title('Pie chart for ' + var, fontsize=25)
plt.savefig('piechart.jpg', dpi = 600)
plt.show()
#Bar chart
applelist[var].plot(kind='bar', fontsize=15,rot=30)
plt.title('Bar chart for ' + var, fontsize=20)
plt.savefig('bargraph.jpg', dpi = 600)
plt.show()
#histogram
applelist[var].plot(kind='hist', fontsize=15,rot=30)
plt.title('Bar chart for ' + var, fontsize=20)
plt.savefig('histogram.jpg', dpi = 600)
plt.show()

#box
filename1 = "brain_data.csv"
f1=open(filename1,"r")
data = pd.read_csv(f1, sep=";", na_values=".")
gender = data.groupby('Gender').sum()
gender.boxplot(column=['FSIQ', 'VIQ','PIQ'])
plt.title("Brain size data")
plt.savefig('boxplot.jpg', dpi = 600)
plt.show()

#scatter
figure = plt.scatter(data['Weight'], data['Height'])
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Weight vs. Height')
plt.savefig('scatterplot.jpg', dpi = 600)
plt.show()

#heatplot
#Generate random data points
x = np.random.randn(100000)
y = np.random.randn(100000)
heatmap, xedges, yedges = np.histogram2d(x, y, bins=(600,600))
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
plt.clf()
plt.imshow(heatmap.T, extent=extent, origin='lower')
plt.show()

#Time series plot
filename = "discoveries.csv"
f = open(filename,'r')
datad = pd.read_csv(f)
datad1 = pd.to_datetime(datad['time'], format = '%Y')
datad2 = datad.set_index('time')
fig = plt.gcf()
fig.set_size_inches(15, 10)
plt.xlabel('Year')
plt.ylabel('Amount of discoveries')
plt.title('Yearly discoveries')
plt.plot(datad2)
plt.savefig('Timeseriesplot.jpg', dpi = 600)
plt.show()
```
